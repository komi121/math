# Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
def length_vector (vec_coordinates : list):
    length_vectors =[]
    for vector in vec_coordinates:
        coordinat = 0
        for coordinats in vector:
            coordinat += coordinats ** 2
        length_vectors.append(coordinat ** 0.5)
    if len(length_vectors) == 1:
        return length_vectors[0]
    else:
        return length_vectors
```

```python
number_vector = int(input("для скольки векторов вы хотите посчитать длину?"))
list_coordinats = []
for vector in range(number_vector):
    list_coordinat = input(f"Введите координаты через пробел для вектора №{vector + 1}").split(" ")
    for index in range(len(list_coordinat)):
        list_coordinat[index] = float(list_coordinat[index])
    list_coordinats.append(list_coordinat)


print(length_vector(list_coordinats))
```

### Окружность

```python
x = []
y = []
y2 = []
x2 = []
R = 1
for i in range(101):
    x_ = i/100
    x.append(x_)
    x2.append(-x_)
    y.append(-(((R ** 2) - x_**2)**0.5))
    y2.append((((R ** 2) - x_**2)**0.5))
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x2,y)
plt.plot(x2,y2)
```

### эллипса

```python
x = []
y = []
y2 = []
x2 = []
R = 1
for i in range(101):
    x_ = i/100
    x.append(x_)
    x2.append(-x_)
    y.append(-((R - x_**2)**0.5) / 4)
    y2.append(((R - x_**2)**0.5) / 4 )
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x2,y)
plt.plot(x2,y2)
```

```python
def polar_to_decart(r, alf):
    alf = alf * (np.pi / 180)
    x = r * np.cos(alf)
    y = r * np.sin(alf)
    return x , y

print(polar_to_decart(5, 45))
```
